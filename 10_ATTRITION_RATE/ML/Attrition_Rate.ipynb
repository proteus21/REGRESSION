```python
tenure_dict=df['Tenure Grp.'].value_counts()
tenure_dict
```

```python
location_dict=df['Location'].value_counts()
location_dict
```

```python
location_new_dict ={
    'Chennai':       7,
    'Noida':         6,
    'Bangalore':     5,
    'Hyderabad':     4,
    'Pune':          3,
    'Mandurai':      2,
    'Lucknow':       1,
    'other place':  0,

}

print( location_new_dict)
```

```python
def location(x):
  if str(x) in location_new_dict.keys():
     return location_new_dict[str(x)]
  else:
    return location_new_dict['other place']
```

```python
df['New Location']=df['Location'].apply(location)
df.head()
```

```python
gen=pd.get_dummies(df['Function'])
```

```python
hr=pd.get_dummies(df['Hiring Source'])
```

```python
# Morital status

def Marital(x):
  if str(x) in Marital_dict.keys() and Marital_dict[str(x)]>100:
    return str(x)
  else:
    return 'other status'
```

```python
df['New Marital']=df['Marital Status'].apply(Marital)
df.head()
```

```python
mr=pd.get_dummies(df["New Marital"])
mr.head()
```

```python
def Promoted(x):
  if x=="Promoted":
    return int(1)
  else:
    return int(0)


df['New Promotion']=df['Promoted/Non Promoted'].apply(Promoted)
df.head()
```

```python
Employee_dic_new ={
    'B1':4,
    'B2':3,
    'B3':2,
    'other group':1,

}

def emp(x):
  if str(x) in Employee_dic_new.keys():
    return str(x)
  else:
    return 'other group'

df['New EMP']=df['Emp. Group'].apply(emp)
df.head()
```

```python
emp=pd.get_dummies(df['New EMP'])
df.head()
```

```python
def Job(x):
  if x=="Yes":
    return int(1)
  else:
    return int(0)


df['New Job Role Match']=df['Job Role Match'].apply(Job)
df.head()
```

```python
jrm=pd.get_dummies(df['New Job Role Match'])
df.head()
```

```python
def Gen(x):
  if x in gender_dict.keys():
    return str(x)
  else:
    return 'other'

df['New Gender']=df['Gender '].apply(Gen)
df.head()
```

```python
gend=pd.get_dummies(df["New Gender"])
tengrp=pd.get_dummies(df['Tenure Grp.'])
data=pd.concat([df,emp, tengrp, gend, mr, hr], axis=1)
data.head()
```

```python
df=pd.read_csv("/content/processing.csv")
df.head()
```

```python
y=df['Stay/Left']
X=df.drop('Stay/Left', axis=1)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=21, stratify =y)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
lr=LogisticRegression(C=0.1, random_state=21, solver='liblinear')
rfc= RandomForestClassifier()
gnb=GaussianNB()
dt=DecisionTreeClassifier()
svc=SVC(C=0.1, random_state=21 )
```

```python
for a, b in zip([lr, rfc, gnb, dt, svc],["Logistic Regression",'Random Forest Classifier', 'Naive Bayes','Dectision Tree','SVC']):
          a.fit(X_train, y_train)
          predict=a.predict(X_train)
          y_pred=a.predict(X_test)
          score_1=accuracy_score(y_train, predict)
          score=accuracy_score(y_test, y_pred)
          print('[%s]training data accurancy is:%f'%(b, score_1))
          print('[%s]testing data accurancy is:%f'%(b, score))



```

```python
#The best model is : Random Forest Classifier
model=rfc.fit(X_train, y_train)
import pickle
with open('Final model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
