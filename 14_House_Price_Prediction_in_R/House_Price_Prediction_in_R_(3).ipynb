##Decision Trees
### Regression

```r
movie_df<-read.csv('/content/Movie_regression.csv', header=TRUE)
str(movie_df)
```

```r
#Data Preprocessing
summary(movie_df)
```

```r
movie_df$Time_taken[is.na(movie_df$Time_taken)]<- mean(movie_df$Time_taken, na.rm=TRUE)
```

```r
#Test_TRAIN SPLIT
set.seed(123)  # Set seed
train_indices <- sample(1:nrow(movie_df), 0.8 * nrow(movie_df))
#Split df onto train and test data
train_set <- movie_df[train_indices, ]
test_set <- movie_df[train_indices, ]
```

```r
# Install library
install.packages("rpart")
install.packages("rpart.plot")
# Load the class package
library(rpart)
library(rpart.plot)
```

```r
#Run regression tree model on train set
regtree <- rpart(formula=Collection~., data=train_set, control=rpart.control(maxdepth = 3))
# Press F1 on rpart for help on this function

# Plot the decision tree
rpart.plot(regtree, box.palette='RdBu', digits=-3)
```

```r
# Predict value at any point
test_set$pred<- predict(regtree, test_set, type='vector')

MSE2_base  <- mean((test_set$pred - test_set$Collection)^2)
print(MSE2_base)
```

```r
#Tree Pruning
fulltree<- rpart(formula=Collection~., data=train_set, control=rpart.control(cp=0))
rpart.plot(fulltree, box.palette = 'RdBu', digits=-3)
printcp(fulltree)
plotcp(regtree)
```

```r
mincp <- regtree$cptable[which.min(regtree$cptable[,'xerror']),'CP']

prunedtree <- prune(regtree, cp=mincp)
rpart.plot(prunedtree, box.palette ="RdBu" , digits=-3)
```

```r
test_set$fulltree <- predict(fulltree, test_set,type="vector")
MSE2_preprun <- mean((test_set$fulltree - test_set$Collection)^2)
print(MSE2_preprun)
```

```r
test_set$pruned <- predict(prunedtree, test_set,type="vector")
MSE2_pruned <- mean((test_set$pruned - test_set$Collection)^2)
print(MSE2_pruned)
```

```r
accuracy_postprun <-mean(test_set$pred == test_set$Collection)
data.frame(MSE2_base, MSE2_preprun, MSE2_pruned)
```

```r
movie_df_class<-read.csv('/content/Movie_classification.csv', header=TRUE)
str(movie_df_class)
```

```r
summary(movie_df_class)
```

```r
movie_df_class$Time_taken[is.na(movie_df_class$Time_taken)]<- mean(movie_df_class$Time_taken, na.rm=TRUE)
#Test_TRAIN SPLIT
set.seed(123)  # Set seed
train_indices <- sample(1:nrow(movie_df_class), 0.8 * nrow(movie_df_class))
#Split df onto train and test data
train_set_c <- movie_df_class[train_indices, ]
test_set_c <- movie_df_class[train_indices, ]
```

```r
# For classification

train_set_c$Start_Tech_Oscar <- as.factor(train_set_c$Start_Tech_Oscar )
test_set_c$Start_Tech_Oscar <- as.factor(test_set_c$Start_Tech_Oscar )
```

```r
# Importing relevant Library e1071
install.packages('e1071')
library(e1071)
```

```r
svmfit =svm(Start_Tech_Oscar~., data=train_set_c, kernel ='linear', cost=1, scala=TRUE)
summary(svmfit)
```

```r
#Predict on test set
ypred=predict(svmfit, test_set_c)
table(predict=ypred, truth=test_set_c$Start_Tech_Oscar)




# Check the support vectors
svmfit$index
```

```r
# Find best value of c /Tuning the hyperparametr

set.seed(123)
tune.out= tune(svm, Start_Tech_Oscar~., data =train_set_c, kernel='linear', ranges=list(coast=c(0.001, 0.01, 0.1, 1, 10, 100)))

bestmod= tune.out$best.model
summary(bestmod)

```

```r
ypredL=predict(bestmod, test_set_c)
table(predict=ypredL, truth=test_set_c$Start_Tech_Oscar)
```

```r
#Polynomial Kernel
svmfitP=svm(Start_Tech_Oscar~., data=test_set_c, kernel='polynomial' , cost=1, degree=2)

#Hyperparametr Tuning

tune.out =tune(svm, Start_Tech_Oscar~., data=train_set_c, cross=4, kernel='polynomial', ranges=list(coast=c(0.001, 0.01, 0.1, 1,5, 10)), degree=c(0.5, 1,2,3,5))
bestmodP=tune.out$best.model
summary(bestmodP)
```

```r
ypredP=predict(bestmodP, test_set_c)
table(predict=ypredP, truth=test_set_c$Start_Tech_Oscar)
```
